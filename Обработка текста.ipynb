```python
f = open('C:\\Users\\79608\\Documents\\txt\\2003.txt', 'r', encoding='utf-8')
text = f.read().lower()
f.close

import re
text = re.sub('\W+',' ', text)
text = re.sub('\d', '', text)
text = re.sub('ё', 'е', text)
#text = re.sub('s p o r t', 's.p.o.r.t.', text)
text = re.sub('рок н ролл', 'рок-н-ролл', text)
text = re.sub('хип хоп', 'хип-хоп', text)
text = re.sub(' би ', ' би-2 ', text)
text = re.sub('ю питер', 'ю-питер', text)

text_out = text.split()

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_w=stopwords.words('russian')+ ['среди', 'немного', 'ради', 'лишь', 'очень', 'сих', 'пор', 'любой', 'сто', 'каждый', 'равно'] + ['эй','м','мм','йе', 'еее', 'ах', 'ууу', 'б', 'ла', 'лалал', 'яя', 'ай', 'яй', 'хэй', 'хей', 'ль', 'ха', 'ка', 'ей', 'э', 'хе', 'ие', 'ко', 'ой', 'лал'] + ['весь', 'оно', 'нем','кого', 'твоей', 'твою', 'тем', 'твой', 'это', 'самый', 'самым', 'самая', 'самое','самой', 'сама', 'сам', 'само', 'собой', 'собою', 'таких', 'всему', 'везде', 'тех', 'зато', 'мою','нашу','нам','тот', 'наши','мной','ту','мое','моих', 'всем', 'мои', 'наших', 'твое', 'тебе', 'чьих', 'своя', 'свой', 'своих', 'свое', 'себя', 'наш', 'сами', 'чей', 'ваш']

text_out = [word for word in text_out if word not in stop_w]
        
import pymorphy2
morph = pymorphy2.MorphAnalyzer()

clean = [morph.parse(word)[0].normal_form for word in text_out]

clean = [word for word in clean if word not in stop_w]

clean_res = ' '.join(clean)
clean_res = re.sub('ё', 'е', clean_res)
clean_res = re.sub('танец минус', 'танцы минус', clean_res)
clean_res = re.sub('ночной снайпер', 'ночные снайперы', clean_res)
clean_res = re.sub('кукрыникс', 'кукрыниксы', clean_res)
clean_res = re.sub('смысловой галлюцинация', 'смысловые галлюцинации', clean_res)
clean_res = re.sub('калиновый мост', 'калинов мост', clean_res)
#clean_res = re.sub('шамбал', 'шамбала', clean_res)
#clean_res = re.sub('гаутаму', 'гаутама', clean_res)
print (clean_res)

f_out = open('C:\\Users\\79608\\Documents\\txt\\clean\\2003.txt','w', encoding='utf-8')
f_out.write(clean_res)
f_out.close
```
